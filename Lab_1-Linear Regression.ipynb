```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
df = pd.read_csv("/content/Real estate.csv")  # Replace with the actual dataset path
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your original dataset

# Create a subset D1 with the first 200 samples
D1 = df.head(200)

# Extract the features X2 to X6 and the target variable Y
features = ['X2', 'X3', 'X4', 'X5', 'X6']
target = 'Y'
subset = D1[features + [target]]

# Plot Y against each of the 5 features separately
for feature in features:
    plt.scatter(subset[feature], subset[target], alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.title(f'{target} vs {feature}')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#importing the datas and relevant libraries

X_train = D1[['X2']]
Y_train = D1['Y']
# Assuming you have X_train and Y_train as pandas DataFrames or arrays

# Initialize weights (u and v), learning rate (LR), and number of iterations (iter)
u = 0
v = 0
LR = 0.01  # Learning Rate
iter = 1000  # Number of iterations

# Custom gradient function
def gradient(X_train, Y_train, u, v):
    N_train = len(Y_train)
    u_gradient = 0
    v_gradient = 0

    for i in range(N_train):
        prediction = np.dot(u, X_train.iloc[i]) + v
        error = prediction - Y_train.iloc[i]
        u_gradient += (-2 / N_train) * error * X_train.iloc[i]
        v_gradient += (-2 / N_train) * error

    return u_gradient, v_gradient

# Custom training function
def train(X_train, Y_train, u, v, LR, iter):
    Loss_his = []

    for _ in range(iter):
        u_gradient, v_gradient = gradient(X_train, Y_train, u, v)
        u -= LR * u_gradient
        v -= LR * v_gradient

        Loss = np.mean((Y_train - (X_train.dot(u) + v)) ** 2)
        Loss_his.append(Loss)

    return u, v, Loss_his

# Convert X_train and Y_train to pandas DataFrames if they are not already
X_train = pd.DataFrame(X_train)
Y_train = pd.Series(Y_train)

# Training using custom functions
u, v, Loss_his = train(X_train, Y_train, u, v, LR, iter)

# Plot the obtained model
plt.scatter(X_train.iloc[:, 0], Y_train, alpha=0.5, label='Data')
plt.plot(X_train.iloc[:, 0], X_train.dot(u) + v, color='red', label=f'LR Model (LR={LR})')
plt.xlabel('X2')
plt.ylabel('Y')
plt.legend()
plt.title('Custom Linear Regression Model')
plt.show()

# Report the final weights and minimum loss
print("Final u (weights):", u)
print("Final v (intercept):", v)
print("Minimum Loss:", Loss_his[-1])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("/content/Real estate.csv")

D1 = data.iloc[:300, :]

def linear_regression(X, weights):
    return np.dot(X, weights)

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def gradient_descent(X, y, weights, learning_rate, num_iterations):
    history = []
    for _ in range(num_iterations):
        y_pred = linear_regression(X, weights)
        error = y_pred - y
        gradient = np.dot(X.T, error) / len(y)
        weights -= learning_rate * gradient
        loss = mean_squared_error(y, y_pred)
        history.append(loss)
    return weights, history

X2 = D1['X2'].values
X3 = D1['X3'].values
X4 = D1['X4'].values
X5 = D1['X5'].values
X6 = D1['X6'].values
Y = D1['Y'].values

X = np.column_stack((X2, X3, X4, X5, X6))
X = np.hstack((np.ones((X.shape[0], 1)), X))
scaler = StandardScaler()
X = scaler.fit_transform(X)

initial_weights = np.zeros(X.shape[1])
learning_rate = 0.001
num_iterations = 100

trained_weights, _ = gradient_descent(X, Y, initial_weights, learning_rate, num_iterations)

y_pred = np.dot(X, trained_weights)

mse = mean_squared_error(Y, y_pred)

print("Mean Squared Error (MSE):", mse)

plt.figure(figsize=(10, 6))
plt.scatter(Y, y_pred, alpha=0.5)
plt.xlabel("Actual Values (Y)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Actual vs. Predicted Values")
plt.show()

# Plot actual vs. predicted values and the fit line
plt.figure(figsize=(10, 6))
plt.scatter(Y, y_pred, alpha=0.5)
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], linestyle='--', color='red', linewidth=2, label='Fit Line')
plt.xlabel("Actual Values (Y)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Actual vs. Predicted Values with Fit Line")
plt.legend()
plt.show()
```
